```python
import scipy.stats as stats # import scipy.stats module as stats to execute the test
```

```python
code = ['A3', 'B3', 'C6', 'D7', 'E6'] #put all selected features as list

#define a small function that helps print pvalues of list of variables when running chi2 tests
def chi(x): #declare function
    for l in x: #set a loop which take as input var list
        chi2_stat, pvalue, dof, ex  = stats.chi2_contingency(df[['gender', l]]) #inside the loop execute the Chi2-test between gender and variable 'l'
        print (str('Chi-2 test between question {} and gender gives as p-value {:.2f}').format(l, pvalue)) #format a print
        
chi(code) #run the function
```
